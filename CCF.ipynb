```python
import numpy as np
import matplotlib.pylab as plt
from scipy import signal
```

```python
x_range = np.arange(0, 10, 0.1)
sin_x = np.sin(x_range)
cos_x = np.cos(x_range)
```

```python
plt.plot(x_range, sin_x)
plt.plot(x_range, cos_x)
```

```python
corr = signal.correlate(sin_x, cos_x)
```

```python
plt.plot(corr)
```

```python
f = np.array([0,0,0,1,1,0])
g = np.array([0,0,0,1,0,0])
```

```python
f_g_convolve =  np.zeros(len(f))

for n in range(-len(g), len(g)):
    for m in range(len(f)):
        fg_sum = 0
        try:
            if (m+n) <0:
                continue
            else:
                fg_sum += f[m]*g[m+n]
        except:
            continue
    f_g_convolve[n] = fg_sum
    
```

```python
plt.plot(f_g_convolve)
```
